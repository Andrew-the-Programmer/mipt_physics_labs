```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import optimize
```

```python
def lin(x, k, b):
    return k * x + b
```

```python
P = np.array([1.0, 1.1, 1.1, 1.2, 1.4, 1.6, 1.7, 1.9, 2.0, 2.2, 2.4, 2.5, 2.7, 2.8, 3.0, 3.1, 3.2, 3.4, 3.5,
              3.7, 3.8, 4.0, 4.1, 4.3, 4.4, 4.5, 4.7, 4.8, 5.0, 5.1, 5.2, 5.4])
#P = np.log(P1 - 0.9)
dt = 2
t = np.arange(0, len(P) *2, 2)
coef, err = optimize.curve_fit(lin, t, P)
x = np.arange(min(t), max(t)+0.02, 0.01)
y = lin(x, coef[0], coef[1])

plt.plot(t, P, 'o')
plt.plot(x, y)
plt.xlabel('t, сек')
plt.ylabel('P, 10^-4 торр')
coef
```
